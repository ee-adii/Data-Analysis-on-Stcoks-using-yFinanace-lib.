```python
class MotifSrch:

    def __init__(self,query_df,series,time_format,df,num_motifs):
     self.Q_df = query_df
     self.time_format = time_format
     self.df = df
     #sereies = typical price column must be in both in query_df and df
     self.series = series
     self.num_motifs = num_motifs
     self.motif_idxs = None
     self.motif_dfs = None

    def srchMotif(self):
     ohlc = {
     'open': 'first',
     'high': 'max',
     'low': 'min',
     'close': 'last',
     'volume': 'sum',
     'typical value':'sum'}
     #converting in same time frame
     new_df = self.df.resample('5min', base=15).apply(ohlc)
     new_query_df = self.Q_df.resample('5min',base=15).apply(ohlc)
     #removing Nan rows
     #new_df =new_df.dropna()
     #new_query_df = new_query_df.dropna()
     #adding a nan row at the end of day
     #for i in range(0,new_df.shape[0]//75):

     #getting the distance matrix
     distance_profile = stumpy.mass(new_query_df['typical value'], new_df['typical value'])
     #taking the top k motifs
     k = self.num_motifs
     idxs = np.argpartition(distance_profile, k)[:k]
     idxs = idxs[np.argsort(distance_profile[idxs])]
     self.motif_idxs = idxs
    # fetching back the OHLCs from the idxs typical value
     nearest_dfs = []
     n = len(self.Q_df)
     for i in range(0,len(idxs)):
       temp_df = self.df[self.motif_idxs[i]:self.motif_idxs[i]+n]
       nearest_dfs.append(temp_df)
     self.motif_dfs = nearest_dfs
     return self.motif_idxs

    def motifsPlot(self,num_plots): 
     if(num_plots<=self.num_motifs):
      for i in range(0,num_plots):
       df_list = self.motif_dfs
       df_list[i]['date'] = df_list[i].index
       df = df_list[i]
       fig = go.Figure(data=[go.Candlestick(x=df['date'],
                open=df['open'], high=df['high'],
                low=df['low'], close=df['close'])
                     ])
       fig.update_layout(xaxis_rangeslider_visible=False)
       fig.show()
      else:
        return "Number of nearest Motif Plots must be less or equal than Total Number of Motifs" 
```

```python
Q_df = df_5min[251:251+5*12]
get_subseq  = MotifSrch(Q_df,'typical value','5min',df_5min,5)
idx = get_subseq.srchMotif()
get_subseq.motifsPlot(5)
idx
```
